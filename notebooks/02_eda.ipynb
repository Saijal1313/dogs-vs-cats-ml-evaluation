```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

google_cleaned_url = "https://raw.githubusercontent.com/Dhairyaxshah/Appfluence/main/data/google_play_cleaned.csv"
apple_cleaned_url  = "https://raw.githubusercontent.com/Dhairyaxshah/Appfluence/main/data/apple_store_cleaned.csv"

df1 = pd.read_csv(google_cleaned_url)
df2 = pd.read_csv(apple_cleaned_url)
```

**DATASET 1 - GOOGLE PLAY STORE**

**EDA Question 1: Market Structure by Category**

This analysis examines how apps are distributed across different categories
to understand market saturation and competition.


```python
df1['Category'].value_counts().head(10)
```

```python
df1['Category'].value_counts().head(10).plot(
    kind='bar',
    figsize=(10,4),
    title='Top 10 Categories by Number of Apps'
)
```

The Family category contains the highest number of apps in the Google Play Store,
followed by Games and Tools. This indicates that the Family category is highly
saturated, suggesting intense competition for visibility and user attention.
As a result, developers entering this category are likely to face greater
competitive pressure compared to less crowded categories.


**EDA QUESTION 2: Are most Google Play apps free or paid?**

```python
df1['Type'].value_counts()
```

```python
df1['Type'].value_counts().plot(
    kind='bar',
    figsize=(5,4),
    title='Free vs Paid Apps on Google Play'
)
```

**Insight:**

The Google Play Store is heavily dominated by free apps, with approximately 7,500 free apps compared to only around 600 paid apps. **EDA Question 3: How is popularity distributed?**

```python
df1['Installs'].describe()
```

```python
df1['log_installs'] = np.log10(df1['Installs'] + 1)


plt.figure(figsize=(8,4))
sns.histplot(df1['log_installs'], bins=30, kde=True)
plt.xlabel('Log10(Installs)')
plt.ylabel('Number of Apps')
plt.title('Distribution of App Installs (Log Scale)')
plt.show()
```

**Insight:**

The distribution of app installs is highly right-skewed, indicating that while most apps receive relatively low install counts, a small number of apps achieve extremely high popularity. Applying a log transformation compresses these extreme values, reduces variance, and makes the distribution more interpretable. This transformation is essential for both analysis and machine learning, as it allows models to learn relative differences in app popularity without being dominated by a few highly downloaded apps.

**EDA Question 4: Do higher-rated apps actually get more installs?**

```python
df1['log_installs'] = np.log10(df1['Installs'] + 1)

plt.figure(figsize=(8,4))
plt.scatter(df1['Rating'], df1['log_installs'], alpha=0.3, s=10)
plt.grid(alpha=0.3)
plt.xlabel('User Rating')
plt.ylabel('Log10(Installs)')
plt.title('Rating vs Log-Scaled Install Counts on Google Play Store')
plt.show()
```

**Insight:**

The scatter plot shows a weak positive relationship between app ratings and install counts. While higher-rated apps are more likely to achieve higher installs, rating alone is not a strong predictor of popularity, as many highly rated apps still have low install counts. This suggests that additional factors such as category, pricing, and visibility play a significant role in app adoption.

**EDA Question 5: Which categories have the highest typical (median) install counts on Google Play?**

```python
df1['log_installs'] = np.log10(df1['Installs'] + 1)

category_median = (
    df1.groupby('Category')['log_installs']
       .median()
       .sort_values(ascending=False)
)
plt.figure(figsize=(10, 5))
category_median.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Median Log10(Installs)')
plt.title('Median App Installs by Category on Google Play')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```

**Insight:**

Categories such as Education, Shopping, and Communication exhibit the highest median install counts, indicating strong typical popularity. However, the similarity in median installs across many top categories suggests intense competition. Categories with slightly lower median installs but fewer competing apps may present more strategic opportunities for new app development.

## Exploratory Data Analysis  App Store (iOS)


This section explores patterns in App Store apps to understand factors influencing app popularity and user ratings.


**EDA Question 1: Which genres dominate the App Store?**

```python
# Count number of apps per genre
genre_counts = (
    df2['prime_genre']
    .value_counts()
    .sort_values(ascending=False)
)

# Plot
plt.figure(figsize=(10, 5))
genre_counts.plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Number of Apps')
plt.title('App Count by Genre on the iOS App Store')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Insight :**

The iOS App Store is overwhelmingly dominated by the Games genre, which contains a significantly higher number of apps than any other category. While genres such as Entertainment and Education follow, their app counts are substantially smaller in comparison. This sharp imbalance indicates extremely high competition within the Games category, whereas many other genres represent comparatively under-served markets with fewer competing apps.

**EDA Question 2: Which genres are the most popular on the App Store based on average (mean) log-scaled rating counts?**

```python
# Log-transform rating counts
df2['log_rating_count'] = np.log10(df2['rating_count_tot'] + 1)

# Compute mean log rating count per genre
genre_popularity = (
    df2.groupby('prime_genre')['log_rating_count']
       .mean()
       .sort_values(ascending=False)
)

# Plot
plt.figure(figsize=(10, 5))
genre_popularity.plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Mean Log10(Rating Count)')
plt.title('Average App Popularity by Genre on the iOS App Store')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Insight :**

Although the Games genre dominates the App Store in terms of app count, it does not rank among the top genres by average popularity. Genres such as Music, Productivity, and Shopping exhibit higher average rating counts, indicating stronger user engagement per app. This suggests that while the Games category is highly saturated, other genres may offer better opportunities for achieving consistent popularity with lower competition.

**EDA Question 2: Does app price affect popularity on the App Store?**

```python
# Log-transform popularity
df2['log_rating_count'] = np.log10(df2['rating_count_tot'] + 1)

plt.figure(figsize=(8, 4))
plt.scatter(df2['price'], df2['log_rating_count'], alpha=0.4)
plt.xlabel('Price (USD)')
plt.ylabel('Log10(Rating Count)')
plt.title('Price vs App Popularity on the iOS App Store')
plt.tight_layout()
plt.show()
```

**Insight :**

The pricepopularity scatter plot shows heavy overlap, with the majority of high-popularity apps clustered at zero price. This indicates that free apps dominate user engagement on the App Store. Paid apps exhibit highly variable popularity with no clear linear relationship between price and rating count. As a result, price alone is a weak continuous predictor of app popularity, motivating the use of a binary free vs paid feature for downstream machine learning models.


**EDA Question 4: Do free apps outperform paid apps in popularity on the App Store?**




```python
# Create free vs paid feature
df2['is_free'] = df2['price'] == 0

# Log-transform popularity (if not already done)
df2['log_rating_count'] = np.log10(df2['rating_count_tot'] + 1)

# Plot
plt.figure(figsize=(6, 4))
sns.boxplot(
    x='is_free',
    y='log_rating_count',
    data=df2
)
plt.xlabel('Is Free App')
plt.ylabel('Log10(Rating Count)')
plt.title('App Popularity: Free vs Paid on the iOS App Store')
plt.tight_layout()
plt.show()
```

**Insight :**

Free apps exhibit a higher median popularity and a wider spread of rating counts compared to paid apps. This indicates that free apps not only attract larger audiences on average but also show greater variability in user engagement. In contrast, paid apps tend to cluster at lower popularity levels with fewer high-engagement outliers. This pattern suggests that free availability is a significant driver of app adoption on the App Store, supporting the use of a binary is_free feature in downstream machine learning models.

**EDA Question 5: Does content rating (age group) influence popularity?**

```python
# Log-transform popularity
df2['log_rating_count'] = np.log10(df2['rating_count_tot'] + 1)

plt.figure(figsize=(6, 4))
sns.boxplot(
    x='cont_rating',
    y='log_rating_count',
    data=df2
)
plt.xlabel('Content Rating (Age Group)')
plt.ylabel('Log10(Rating Count)')
plt.title('App Popularity by Content Rating on the iOS App Store')
plt.tight_layout()
plt.show()
```

**Insight :**

Apps targeting broader age groups (4+ and 12+) tend to achieve higher median popularity, while more restrictive content (17+) shows lower and more variable engagement. This indicates that content rating meaningfully influences audience reach and should be retained as a categorical feature for prediction and recommendation tasks.
